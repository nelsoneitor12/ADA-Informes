#2. Insertion Sort

Veremos una implementacion del algoritmo **InsertionSort** en el siguiente codigo:

##2.1. Codigo

```python
import random
from termcolor import colored
import copy
from timeit import default_timer

def Insertion_Sort(arr, verbose=False):
  
  tam=len(arr)
  comp=0 
  cont=0
  auxi=0
  if verbose == True: print("input array:",arr)
  inicio=default_timer()
  for i in range(1,tam+1):
    if verbose == True: print("\nPasada",i)
    all_sort = True
    if i==tam : break
    for j in range(i, 0, -1):
      comp+=1
      all_sort=False
      if arr[j]<arr[j-1]:
          arr[j] , arr[j-1] = arr[j-1] , arr[j]
          auxi+=1 
          if verbose == True: 
             print(str(arr[:j-1])[1:-1],",",colored(f"{str(arr[j-1:j+1])[1:-1]}","red"),",",str(arr[j+1:])[1:-1])
      
      else:
        break

    if all_sort == False and verbose == True: 
           print("\nAl finalizar pasada:")
           print(arr)
  if all_sort: 
    if verbose == True: 
      print("is sorted!\n")
      print(f"output array: {arr}")
  final=default_timer()
  tiempo=final-inicio
  auxi=auxi/(i-1)
  return arr, comp, auxi, tiempo

# Ejemplo
A = [3,5,4,8,6,1]
print("Entrada: ",A)
A, counter, aux, tiemp = Insertion_Sort(A)
print("Salida: ",A)
print("Media de retrocesos: ", aux)
print("# comparaciones: ", counter)
print(f"tiemp: {tiemp} seg")
```

##2.2 Descripcion del Algoritmo



El algoritmo tiene como entrada el arreglo "arr" con la secuencia de tamaño "tam", de la cual sus elementos se iran ordenando en el mismo arreglo.
Los pasos seran:
1. Durante cada iteracion i una sub-iteracion j, la cual comienza con valor de indice i=j y en primera instancia con i=j=1 ira comparando el elemento arr[j] con arr[j-1], restando una unidad al indice j en cada iteracion hasta llegar a 0. Si arr[j]<arr[j-1] se intercambian los elementos.
2. Si el elemento anterior a arr[j] no es mayor se quiebra la iteracion j, y se avanza al i siguiente hasta llegar al final del arreglo.
3. Cuando se llegue el ultimo elemento del arreglo y se compare con el/los elementos anteriores dependiendo del caso. Se dara por ordenado el arreglo.
4. Cuando se finalize, se retornara el arreglo ordenado con un contador de operaciones, la media de las veces que se retrocedieron los elementos, y el tiempo que se demoro en ordenar los elementos.

A diferencia de en otros algoritmos de ordenamiento, en este es necesario llegar hasta el final de la secuencia. Ya que se va comparando cada elemento con sus respectivos antecesores para retroceder cuando corresponda.

# 3. Tiempo de ejecución

### **Teorema (Tiempo de ejecución).**

*En el peor caso el algoritmo **InsertionSort** tiene un **tiempo de ejecución de** $O(n^2)$.*

## Prueba del teorema

Podemos apreciar viendo como opera el algoritmo, que el **peor caso** ocurre cuando la secuencia se encuentra en el orden inverso al correcto *(Ej: [4,3,2,1])*. En este caso, en cada pasada el elemento se movería una **n-i** casillas a partir del segundo elemento.

### Ejemplo

```python
a =[5,4,3,2,1];
print("Entrada sin ordenar:",a)
a,counter, mid, time=Insertion_Sort(a,True)
print("Salida ordenada:",a)
print("Total de comparaciones realizadas:",counter)
print("Media de retrocesos:", mid)
```

Se puede ver claramente como en cada pasada se hace una comparacion mas que la anterior, obtenion 10 comparaciones y una media de 2.5 retrocesos por pasada.

En el caso promedio, se deberían realizar $\sum\limits_{i=0}^{n-1} i={n(n-1)}$ comparaciones. Con esto confimamos que el tiempo de ejecucion en el peor caso es de $O(n^2)$.

# 4. Correctitud

### **Teorema (Correctitud).**

*El algoritmo **InsertioneSort** genera un arreglo: **[a_1',a_2',...,a_n']**, con los mismos elementos del arreglo de entrada ordenados de menor a mayor, es decir,* $a_1'\leq a_2' \leq... \leq a_n'$.


## Prueba del Teorema

Pudimos observar en los ejemplos anteriores, que en cada iteracion si el elemento anterior al arr[j] era menor estos se intercambiaban.

Con esto obtenemos la siguiente **propiedad invariante de bucle** para el el algoritmo `InsertionSort`:

> Al comienzo de cada iteración $i$ del bucle más grande,los últimos $i-1$ elementos del arreglo corresponden a los mayores elementos del arreglo original ordenados de menor a mayor.

**Inicialización**

Al comienzo de la primera iteración (i=1), los últimos $i-1=0$ elementos del arreglo cumplen la propiedad. Esto es trivialmente cierto.

**Mantención**

Aplicando inducción, asumimos que al comienzo de cada iteración $i$ se cumple la propiedad. Es decir, los últimos $i-1$ elementos del arreglo corresponden a los mayores elementos del arreglo original ordenados de menor a mayor.

```python
import matplotlib.pyplot as plt

x=[n for n in range(5,20)] 
y=[];
y1=[n*(n-1) for n in range(5,20)]
y2=[n-1 for n in range(5,20)]

for n in range(5,20):
  a = random.sample(range(1, 20), n)
  a,counter, mid, tiemp = Insertion_Sort(a)
  y.append(counter)
  

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Insertion Sort", "Peor Caso", "Mejor Caso"])

plt.xlabel('n')
plt.ylabel('numero de operaciones')
plt.show()
```

```python
def bubble_sort(a, verbose=False):
    n = len(a)
    T = 0 #contador de comparaciones

    if verbose == True: print("input array:",a)
    inicio=default_timer()
    for i in range(1,n+1):
        if verbose == True: print("\nPasada",i)
        # Flag que indica si el arreglo ya se encuentra ordenado
        all_sort = True
        # Se recorre el arreglo
        for j in range(n - i):
            T +=1
            if a[j] > a[j + 1]:
                # Si el elemento es mayor al siguiente se intercambia
                a[j], a[j + 1] = a[j + 1], a[j]
                if verbose == True: 
                  print(str(a[:j])[1:-1],",",colored(f"{str(a[j:j+2])[1:-1]}","red"),",",str(a[j+2:])[1:-1])

                all_sort = False

        if all_sort == False and verbose == True: 
           print("\nAl finalizar pasada:")
           print(str(a[:n-i])[1:-1],",",colored(f"{str(a[n-i:])[1:-1]}","blue"))

        if all_sort:
            final=default_timer()
            if verbose == True: print("is sorted!")
            break

    if verbose == True: print("\noutput array:",a)
    tiempo=final-inicio
    return a, T, tiempo

# Ejemplo
A = [5,4,3,2,1]
print("Entrada: ",A)
A, counter, tiemp = bubble_sort(A)
print("tiempo=", tiemp)
print("Salida: ",A)
print("# comparaciones: ", counter)
```
